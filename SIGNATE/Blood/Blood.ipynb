```python
def kesson_table(df): 
        null_val = df.isnull().sum()
        percent = 100 * df.isnull().sum()/len(df)
        kesson_table = pd.concat([null_val, percent], axis=1)
        kesson_table_ren_columns = kesson_table.rename(
        columns = {0 : '欠損数', 1 : '%'})
        return kesson_table_ren_columns
 
display(kesson_table(train))
display(kesson_table(test))
 
```

```python
# scikit-learnのインポートをします
from sklearn import tree
```

```python
# 「train」の目的変数と説明変数の値を取得
target = train["target"].values
features_one = train[["age","anaemia","creatinine_phosphokinase","diabetes","ejection_fraction","high_blood_pressure","platelets","serum_creatinine","serum_sodium",	"sex",	"smoking",	"time"	]].values
 
# 決定木の作成
my_tree_one = tree.DecisionTreeClassifier()
result = my_tree_one
my_tree_one = my_tree_one.fit(features_one, target)

 
# 「test」の説明変数の値を取得
test_features = test[["age","anaemia","creatinine_phosphokinase","diabetes","ejection_fraction","high_blood_pressure","platelets","serum_creatinine","serum_sodium",	"sex",	"smoking",	"time"]].values
 
# 「test」の説明変数を使って「my_tree_one」のモデルで予測
my_prediction = my_tree_one.predict(test_features)
result.score(features_one,target)
```

```python
# PassengerIdを取得
index = np.array(test["id"]).astype(int)
 
# my_prediction(予測データ）とPassengerIdをデータフレームへ落とし込む
my_solution = pd.DataFrame({'id': index, 'target': my_prediction})
# my_tree_one.csvとして書き出し
my_solution.to_csv("my_tree_one.csv",header=False,index=False)
```
