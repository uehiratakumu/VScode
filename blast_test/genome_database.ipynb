```python
from Bio.Blast import Applications
import multiprocessing
from Bio.Blast import NCBIXML
core = multiprocessing.cpu_count()
sost = "/Users/takumu/VScode/signal_db/sost_gene.fna"
cline = Applications.NcbiblastnCommandline(query=sost, db=path, out='blastn-out_sost.xml', outfmt=5, num_threads=core)
stdout, stderr = cline()
# 出力ファイルの読み込み

with open('blastn-out_sost.xml', 'rt') as handle:
    blast_result = NCBIXML.read(handle)
descriptions = blast_result.descriptions
     
for description in descriptions:
    print(description.title)
    print('スコア値(raw)：' + str(description.score))
    print('スコア値(bit)：' + str(description.bits))
    print('E-value：' + str(description.e))
    print('HSPの個数：' + str(description.num_alignments))
    print('')
    
alignments = blast_result.alignments
alignments[0].hsps[0].sbjct_end    
```

```python
from Bio import TogoWS, SeqIO
with TogoWS.entry('nucleotide', '563318386','fasta') as handle:
    record = SeqIO.read(handle, 'fasta')
    SeqIO.write(record, 'seq_1.fasta', 'fasta')
```

```python
# FASTAファイルを読み込む
file_path = "/Users/takumu/VScode/Mus_musculus_db/Mus_musculus.GRCm39.ncrna.fa"
records = list(SeqIO.parse(file_path, "fasta"))

# 各エントリに関する情報を取得する
rna_info = []
for record in records:
    x = list(record.description.split(" "))
    y = []
    for i in range(len(x)):
        z = x[i].split(":")
        for j in range(len(z)):
                y.append(z[j])
    z = []
    for j in range(len(y)):
        if (j == 0) or (j == 4) or (j == 5) or (j == 6) or (j == 7) or (j == 9) or (j == 11):
            z.append(y[j])                
    z.append(str(record.seq))
    rna_info.append(z)

# 先頭5つのエントリに関する情報を表示
for info in rna_info[:5]:
    print(info)
```
